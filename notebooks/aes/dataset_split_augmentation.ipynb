```python
import random
```

```python
def print_feedback_essays(tokens, prompt, seed):
    output = llm.create_chat_completion(
        messages = [
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_tokens=tokens,
        temperature=random.uniform(0.7, 0.8),
        top_p=random.uniform(0.9, 0.99),

        seed=seed
            )['choices'][0]['message']['content']
    output = output.split('/think')[1]
    print(output, "\n\n\n\n")
    return {'full_text': output,
            'Overall':1.0,
            'Cohesion':1.0,
            'Syntax':1.0,
            'Vocabulary':1.0,
            'Phraseology':1.0,
            'Grammar':1.0,
            'Conventions':1.0}
```

```python
generated_essays = []
```

```python
tokens = 400
for topic in what_to_gen.keys():
    prompt = prompt_ordinary
    if topic in ['Community service', 'Grades for extracurricular activities', 'Cell phones at school']:
        tokens = 200
    if topic == 'Distance learning':
        prompt = prompt_distance
    if topic == 'Being busy':
        prompt = prompt_busy
    prompt = prompt.format(topic)
    for seed in range(int(what_to_gen[topic])):
        generated_essays.append(print_feedback_essays(tokens, prompt, seed))
```

```python
df_generated = pd.DataFrame(generated_essays)
```
